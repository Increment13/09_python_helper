```python
import configparser

#lib for connect to sharepoint
from office365.runtime.auth.user_credential import UserCredential
from office365.sharepoint.client_context import ClientContext
from office365.sharepoint.files.file import File 

#lib for read copy file from shared 
import io

#pandas %)
import pandas as pd
```

```python
#read config(do not store passwords in the public) 
config = configparser.ConfigParser()
config.read('../config.ini')

user_share = config.get('SHARE_POINT', 'user')
password_share = config.get('SHARE_POINT', 'password')
sharepoint_url = config.get('SHARE_POINT', 'url')
sharepoint_doc = config.get('SHARE_POINT', 'doc_library')

user_credentials = UserCredential(user_share, password_share)
ctx = ClientContext(sharepoint_url).with_credentials(user_credentials)
```

```python
#excample path and file
folder_name = 'Creditportfolio/2023/20230531'
file_name = 'WL 2023-05-31.xlsb'

file_url = "/".join([sharepoint_doc,folder_name,file_name])

response = File.open_binary(ctx, file_url)
#if you get something other than a 200 response, you should check the path or file name 
print(response)

#save data to BytesIO stream
bytes_file_obj = io.BytesIO()
bytes_file_obj.write(response.content)
bytes_file_obj.seek(0) #set file object to start
```

```python
#here read exel, csv or else
df = pd.read_excel(bytes_file_obj, sheet_name='31.05.2023')
df.head()
```
